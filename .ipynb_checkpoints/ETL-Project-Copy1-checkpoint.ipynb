```python
# Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

# Child vaccination rates

```python
# File to Load
Child_vacc_rates = "Resources/Child_vacc_rates.csv"

# Read the modified GoodReads csv and store into Pandas DataFrame
Child_vacc_rates = pd.read_csv(Child_vacc_rates, encoding="utf-8")
Child_vacc_rates.head()
```

```python
new_Child_vacc_rates = Child_vacc_rates[['LOCATION','INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'TIME', 'Value']]
new_Child_vacc_rates.head()
```

```python
new_Child_vacc_rates.count()
```

```python
new_Child_vacc_rates = new_Child_vacc_rates.dropna(how="any")
new_Child_vacc_rates.count()
```

```python
new_Child_vacc_rates.describe()
```

```python
selected_new_Child_vacc_rates = new_Child_vacc_rates.loc[(new_Child_vacc_rates["LOCATION"] == "JPN") | (new_Child_vacc_rates["LOCATION"] == "CAN")]
selected_new_Child_vacc_rates
```

```python
measles_new_Child_vacc_rates = selected_new_Child_vacc_rates.loc[(selected_new_Child_vacc_rates["SUBJECT"] == "MEASLES") & (selected_new_Child_vacc_rates["TIME"] >= 2000)]
measles_new_Child_vacc_rates
```

# Influenza vaccination rates

```python
# File to Load
Influenza_vacc_rates = "Resources/Influenza_vacc_rates.csv"

# Read the modified GoodReads csv and store into Pandas DataFrame
Influenza_vacc_rates = pd.read_csv(Influenza_vacc_rates, encoding="utf-8")
Influenza_vacc_rates.head()
```

```python
new_Influenza_vacc_rates = Influenza_vacc_rates[['LOCATION','INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'TIME', 'Value']]
new_Influenza_vacc_rates.head()
```

```python
new_Influenza_vacc_rates.count()
```

```python
new_Influenza_vacc_rates = new_Influenza_vacc_rates.dropna(how="any")
new_Influenza_vacc_rates.count()
```

```python
new_Influenza_vacc_rates.describe()
```

```python
selected_new_Influenza_vacc_rates = new_Influenza_vacc_rates.loc[(new_Influenza_vacc_rates["LOCATION"] == "JPN") | (new_Influenza_vacc_rates["LOCATION"] == "CAN")]
selected_new_Influenza_vacc_rates
```

# Merging

```python
merge_df = pd.merge(measles_new_Child_vacc_rates, selected_new_Influenza_vacc_rates, on=["LOCATION", "TIME"], how="inner", suffixes=("_Child", "_Flu"))
merge_df.head()
```

```python
new_merge_df = merge_df[['LOCATION','INDICATOR_Child', 'SUBJECT_Child', 'MEASURE_Child', 'TIME', 'Value_Child', 'INDICATOR_Flu', 'SUBJECT_Flu', 'MEASURE_Flu', 'Value_Flu']]
new_merge_df.head()
```

# Export to SQL

```python
connection_string = "postgres:password@localhost:5432/Vaccines" 
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
new_merge_df.to_sql(name='vaccination_rates', con=engine, if_exists='append', index=False)
```
